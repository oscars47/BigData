```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
D = np.array([[0, 1], [2, 3], [3, 6], [4, 8]])
# plot the data and the line
A = np.array([[1, 0], [1, 2], [1, 3], [1, 4]])
b = np.array([1, 3, 6, 8])
theta = np.array([[18/35], [63/35]])

plt.plot(D[:, 0], D[:, 1], 'ro', label='Data')
plt.plot(D[:, 0], A.dot(theta), 'b', label='Line')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig('2c.png')
```

```python
# generate 100 points near the line from gaussian distribution
np.random.seed(0)
# generate 100 points along the line; then add gaussian noise
points = np.linspace(0, 5, 100)
# evaluate the line at the points
def line(x): return theta[0] + theta[1] * x
# generate gaussian noise
noise = np.random.normal(0, 1, 100)
points = line(points) + noise

# now plot the points
plt.figure()
plt.plot(D[:, 0], D[:, 1], 'ro', label='Data')
plt.plot(D[:, 0], A.dot(theta), 'b', label='Line')
plt.plot(np.linspace(0, 5, 100), points, 'go', label='Generated Data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig('2d.png')
```
